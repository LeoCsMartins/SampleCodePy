```python
# Função para formatação do arquivo json (fornecida pelo professor)

def from_lists_of_dicts(column):
    """
    Cria um pandas.DataFrame a partir de 
     uma lista de listas de dicionários
    
    Parametros:
    -----------
        column: pandas.Series
        
    Retorno:
    --------
        pandas.DataFrame
    """
    is_list = column.apply(lambda x: isinstance(x, list))
    only_lists = column[is_list]

    # only_lists.sum() concatena todas as listas no pandas.Series
    df = pd.DataFrame.from_dict(only_lists.sum())

    # criando uma coluna para mapear as novas linhas
    # ao perfil correto no json
    df['perfil_id'] = np.repeat(only_lists.index,
                                   only_lists.apply(lambda x: len(x)))

    return df
    
from_lists_of_dicts(df.languages)
```

```python
# Pesquisa lista de idiomas falados por membros da equipe, visando identificar habilidades e competencias da equipe.

language = from_lists_of_dicts(df.languages) # cria variável que recebe data frame de linguagens

#Método para extrair apenas a especificação do idioma falado, armazendo em um vetor  
language['idiom'] = pd.Series((n.split()[1] for n in language.name), index=language.index) 

print(language)
```

```python
# Bloco criado para ler o nome dos arquivos csv. Os nomes serão inseridos no data frame de cada curso, o que permitirá a 
# unificar todos os arquivos em um data frame

from os import walk 

f = []
for (dirpath, dirnames, filenames) in walk("C:/Users/Leonardo Martins/Desktop/ILE/Trabalho/rankings"):
    f.extend(filenames)
    break
name_of_courses = filenames = ' '.join(filenames).replace('.csv','').split()
    
print(name_of_courses)
```

```python
# Lê todos os arquivos a serem importados presentes no diretorio especificado

import glob

concat_files = glob.glob('C:/Users/Leonardo Martins/Desktop/ILE/Trabalho/rankings/*.csv')
print (concat_files)
```

```python
# Automatiza a importação de todos os arquivos de avaliação dos cursos (39) presentes em diretório 

list_concat_df = []

for csv_filename in concat_files:
    df = pd.read_csv(csv_filename, sep=';')
    df['Nome do Curso'] = name_of_courses[len(list_concat_df)] # Insere o nome do curso em coluna do data frame
    list_concat_df.append(df)

print(len(list_concat_df)) #certicar a quantidade de nomes dos arqquivos inseridos no dataframe
```

```python
# Instancia uma variável que recebe a lista de cursos concatenados

group_courses = pd.concat(list_concat_df)
print(group_courses)
```
